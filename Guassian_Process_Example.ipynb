```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

# Generate some training data
X = np.linspace(-5, 5, 10).reshape(-1, 1)
y = np.sin(X).ravel() + np.random.normal(0, 0.2, X.shape[0])  # Noisy sine wave

# Define the kernel as a combination of a constant and an RBF kernel
kernel = C(1.0) * RBF(length_scale=1.0)

# Create the Gaussian Process Regressor
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)

# Fit to training data
gp.fit(X, y)

# Generate test points
X_test = np.linspace(-6, 6, 100).reshape(-1, 1)

# Predict using the trained GP model
y_pred, sigma = gp.predict(X_test, return_std=True)

# Plot results
plt.figure(figsize=(8, 5))
plt.plot(X, y, 'ro', label="Training data")
plt.plot(X_test, y_pred, 'b-', label="Mean prediction")
plt.fill_between(X_test.ravel(), y_pred - 1.96 * sigma, y_pred + 1.96 * sigma, alpha=0.2, label="Confidence interval")
plt.legend()
plt.title("Gaussian Process Regression")
plt.show()
```
